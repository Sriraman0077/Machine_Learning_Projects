```python
import pandas as pd
df=pd.read_excel('/content/Underweight orgi.xlsx')
df
```

```python
#Splitting the dataset into two underweight and normal classes
import matplotlib.pyplot as plt
u_df=df[df["Class"]=="Underweight"]
n_df=df[df["Class"]=="Normal"]
plt.scatter(u_df['Weight'],u_df['Height'],c='r',label="Underweight")
plt.scatter(n_df['Weight'],n_df['Height'],c='b',label="Normal")
plt.scatter(57,170,s=70,c='y') #Test data to differentiate
plt.legend()
plt.show()
```

```python
#Initializing X and Y for predicting the model so X would be Weight and Height and Y would be class for this we are gonna use ILOC
X=df.iloc[:,[0,1]].values     # :=I want all rows and [0,1] columns that are required and .values to convert series to numpy array
y=df.iloc[:,[2]].values
```

```python
#Training the model
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
#Testing the model
test_data=(57,170)
neigh.predict([test_data])
```

```python
#Using neighbours method
d,i=neigh.kneighbors([test_data])
```

```python
#Validating the values which we got in excel
print(d)
print(i)
```
